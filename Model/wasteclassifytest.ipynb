```python
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = load_model('waste_model.h5')
```

```python
model.summary()
```

```python
def loadimg(path):
    img = cv2.imread(path)
    img = cv2.resize(img,(224,224))
    return img.reshape(1, 224, 224, 3)

categories = ['Electronic Waste', 'Glass Waste', 'Metal Waste', 'Organic Waste', 'Paper Waste', 'Plastic Waste', 'Textile Waste']
```

```python
test_img = loadimg('../Test/718sn7oOcfL.jpg')
test_img.shape
plt.imshow(test_img[0])
result = np.argmax(model.predict(test_img), axis = 1)
print("Predicted: " + categories[result[0]])
```
